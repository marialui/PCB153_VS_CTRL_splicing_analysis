```python
import math

def bed_length_stats(bed_file):
    lengths = []

    with open(bed_file) as f:
        for line in f:
            if line.startswith("#") or not line.strip():
                continue
            chrom, start, end, *rest = line.rstrip().split()
            lengths.append(int(end) - int(start))

    n = len(lengths)
    mean = sum(lengths) / n
    variance = sum((x - mean) ** 2 for x in lengths) / n
    std = math.sqrt(variance)

    return n, mean, std


# ---- BED file paths ----
background_n = "/home/maria/Desktop/progetto_inquinanti/risultati_inquinanti/PCB153-vs-CTR/RBP-DEGS/RBPbench/RBP_DEG_specific/intellectual_disability_negative_PSI_PCB153-vs-CTR_ASE_specific_coordinates_enmo/background_sequences.tmp.bed"
background_p = "/home/maria/Desktop/progetto_inquinanti/risultati_inquinanti/PCB153-vs-CTR/RBP-DEGS/RBPbench/RBP_DEG_specific/intellectual_disability_positive_PSI_PCB153-vs-CTR_ASE_specific_coordinates_enmo/background_sequences.tmp.bed"

dase_not_expanded_negative = "/home/maria/Desktop/progetto_inquinanti/risultati_inquinanti/PCB153-vs-CTR/miRNA/negative_PSI/intellectual_disability_negative_PSI_PCB153-vs-CTR_ASE_specific_coordinates.bed"
dase_not_expanded_positive='/home/maria/Desktop/progetto_inquinanti/risultati_inquinanti/PCB153-vs-CTR/miRNA/positive_PSI/intellectual_disability_positive_PSI_PCB153-vs-CTR_ASE_specific_coordinates.bed'


# ---- compute stats ----
n_bg_n, mean_bg_n, std_bg_n = bed_length_stats(background_n)
print("background negatives")
print(f"  intervals: {n_bg_n}")
print(f"  mean length: {mean_bg_n:.2f}")
print(f"  std dev: {std_bg_n:.2f}\n")

# ---- compute stats ----
n_bg_p, mean_bg_p, std_bg_p = bed_length_stats(background_p)
print("background positives")
print(f"  intervals: {n_bg_p}")
print(f"  mean length: {mean_bg_p:.2f}")
print(f"  std dev: {std_bg_p:.2f}\n")

n_da_n, mean_da_n, std_da_n = bed_length_stats(dase_not_expanded_negative)
mean_da_negative_plus_500 = mean_da_n + 500

print("Dase expanded negative")
print(f"  intervals: {n_da_n}")
print(f"  mean length : {mean_da_negative_plus_500:.2f}")
print(f"  std dev: {std_da_n:.2f}\n")

n_da_p, mean_da_p, std_da_p = bed_length_stats(dase_not_expanded_positive)
mean_da_positive_plus_500 = mean_da_p + 500

print("Dase expanded positive")
print(f"  intervals: {n_da_p}")
print(f"  mean length : {mean_da_positive_plus_500:.2f}")
print(f"  std dev: {std_da_p:.2f}\n")
```

```python
import pandas as pd

# Read the two dataframes
df_pos = pd.read_csv("RBP_DEG_specific_enriched_motif_DASE_positive_psi_intellectual_disability.csv")
df_neg = pd.read_csv("RBP_DEG_specific_enriched_motif_DASE_negative_psi_intellectual_disability.csv")

# Find common columns
common_cols = df_pos.columns.intersection(df_neg.columns)

# Subset to common columns
df_pos_common = df_pos[common_cols]
df_neg_common = df_neg[common_cols]

# Concatenate
df_combined = pd.concat([df_pos_common, df_neg_common], ignore_index=True)
# Remove unwanted columns if they exist
df_combined = df_combined.drop(columns=["IncLevel1", "IncLevel2"], errors="ignore")

# Optional: save to file
# df_combined.to_csv("enriched_motif_DASE_combined.csv", index=False)
df_combined.head()
```

```python
import pandas as pd
import re

def determine_motif_position_strand(row):
    # Use regex to extract coordinates and strand
    match = re.match(r'(.+):(\d+)-(\d+)\((\+|\-)\)', row['expanded_sequence'])
    if not match:
        # If format is unexpected, mark as unknown
        return 'unknown'
    
    chrom, seq_start, seq_end, strand = match.groups()
    seq_start = int(seq_start)
    seq_end = int(seq_end)
    gen_s = int(row['gen_s'])
    gen_e = int(row['gen_e'])

    # Define upstream/downstream regions based on strand
    if strand == '+':
        upstream_start = seq_start
        upstream_end = seq_start + 250
        downstream_start = seq_end - 250
        downstream_end = seq_end
    else:  # strand == '-'
        upstream_start = seq_end - 250
        upstream_end = seq_end
        downstream_start = seq_start
        downstream_end = seq_start + 250

    # Check if gen_s-gen_e falls in upstream/downstream
    if gen_e <= upstream_end and gen_s >= upstream_start:
        return 'upstream'
    elif gen_s >= downstream_start and gen_e <= downstream_end:
        return 'downstream'
    else:
        return 'exon'

# Apply to dataframe
df_combined['motif_position'] = df_combined.apply(determine_motif_position_strand, axis=1)
```
