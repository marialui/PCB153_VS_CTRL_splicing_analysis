```python
comp1='rMATS_PCB153-vs-CTR/'
comp2='rMATS_PCB118-vs-CTR/'
comp3='rMATS_PCB153-Tau-vs-CTR-Tau/'
comp4='rMATS_PCB118-Tau-vs-CTR-Tau/'
comp5='rMATS_CTR-Tau-vs-CTR/'
comp6='rMATS_bAmy-vs-CTR/'
```

```python
LIST_EVENTS=['A3SS', 'A5SS','MXE', 'RI', 'SE']
```

```python
def calculate_average(row):
    ijc1 = list(map(int, row['IJC_SAMPLE_1'].split(',')))
    sjc1 = list(map(int, row['SJC_SAMPLE_1'].split(',')))
    ijc2 = list(map(int, row['IJC_SAMPLE_2'].split(',')))
    sjc2 = list(map(int, row['SJC_SAMPLE_2'].split(',')))
    all_values = ijc1 + sjc1 + ijc2 + sjc2
    return sum(all_values) / len(all_values)
```

```python
for ae in LIST_EVENTS:
    PCB153_vs_CTR=pd.read_csv(comp1+'%s.MATS.JC.txt' %(ae),sep='\t')
    # Apply the function to each row
    PCB153_vs_CTR['Mean coverage'] = PCB153_vs_CTR.apply(calculate_average, axis=1)
    original_PCB153_vs_CTR=PCB153_vs_CTR
    print(ae, PCB153_vs_CTR.columns)
    PCB153_vs_CTR=PCB153_vs_CTR[PCB153_vs_CTR['FDR'] < 0.05]
    PCB153_vs_CTR = PCB153_vs_CTR[(PCB153_vs_CTR['IncLevelDifference'] > 0.1) | (PCB153_vs_CTR['IncLevelDifference'] < -0.1)]
    PCB153_vs_CTR=PCB153_vs_CTR[PCB153_vs_CTR['Mean coverage'] >5]
    PCB153_vs_CTR.to_csv('%s/DE_novo_SIG-PCB153-vs-CTR-%s-JC.txt'%(ae,ae),sep='\t')
    print(PCB153_vs_CTR.shape,original_PCB153_vs_CTR.shape)
    diff_df_PCB153_vs_CTR = original_PCB153_vs_CTR.merge(PCB153_vs_CTR, how='left', indicator=True)
    # Select rows that are only present in original_bAmy_VS_CTR
    not_sig_PCB153_vs_CTR= diff_df_PCB153_vs_CTR[diff_df_PCB153_vs_CTR['_merge'] == 'left_only'].drop(columns=['_merge'])
    not_sig_PCB153_vs_CTR.to_csv('%s/DE_novo_NOT-SIG-PCB153-vs-CTR-%s-JC.txt'%(ae,ae),sep='\t')
```

```python
for ae in LIST_EVENTS:
    # Load PCB153-Tau vs CTR-Tau file
    PCB153_Tau_vs_CTR_Tau = pd.read_csv(comp3 + '/%s.MATS.JC.txt' % ae, sep='\t')

    # Compute mean coverage
    PCB153_Tau_vs_CTR_Tau['Mean coverage'] = PCB153_Tau_vs_CTR_Tau.apply(calculate_average, axis=1)

    # Save original table
    original_PCB153_Tau_vs_CTR_Tau = PCB153_Tau_vs_CTR_Tau.copy()

    print(ae, PCB153_Tau_vs_CTR_Tau.columns)

    # Apply filters
    PCB153_Tau_vs_CTR_Tau = PCB153_Tau_vs_CTR_Tau[PCB153_Tau_vs_CTR_Tau['FDR'] < 0.05]
    PCB153_Tau_vs_CTR_Tau = PCB153_Tau_vs_CTR_Tau[
        (PCB153_Tau_vs_CTR_Tau['IncLevelDifference'] > 0.1) |
        (PCB153_Tau_vs_CTR_Tau['IncLevelDifference'] < -0.1)
    ]
    PCB153_Tau_vs_CTR_Tau = PCB153_Tau_vs_CTR_Tau[PCB153_Tau_vs_CTR_Tau['Mean coverage'] > 5]

    # Save significant events
    PCB153_Tau_vs_CTR_Tau.to_csv(
        '%s/DE_novo_SIG-PCB153-Tau-vs-CTR-Tau-%s-JC.txt' % (ae, ae),
        sep='\t'
    )

    print(PCB153_Tau_vs_CTR_Tau.shape, PCB153_Tau_vs_CTR_Tau.shape)

    # Find non-significant events
    diff_df = original_PCB153_Tau_vs_CTR_Tau.merge(
        PCB153_Tau_vs_CTR_Tau, how='left', indicator=True
    )
    not_sig = diff_df[diff_df['_merge'] == 'left_only'].drop(columns=['_merge'])

    # Save non-significant events
    not_sig.to_csv(
        '%s/DE_novo_NOT-SIG-PCB153-Tau-vs-CTR-Tau-%s-JC.txt' % (ae, ae),
        sep='\t'
    )
```
