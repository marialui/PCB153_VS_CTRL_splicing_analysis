```python
for comp in [comp1, comp2,comp3,comp4,comp5,comp6]:
    print(comp)
    path_to_search = '**/DE_novo_SIG-%s*-JC.txt' %(comp)  # This pattern searches for CSV files in all subdirectories
    csv_files = glob.glob(path_to_search, recursive=True)
    print(path_to_search,csv_files)
    for ev in csv_files:
        print(ev)
    # Read the file while skipping the first two columns
        df_tmp = pd.read_csv(ev, sep='\t', usecols=lambda column: column not in [ 'Unnamed: 0','ID'])
        print(df_tmp.columns[0:12])
        #print(df_tmp.head())
        clas=ev.split('/')[0]
        #print(clas)
        df_tmp['Class']=clas
        print('<<<<<',clas)
        if clas=='MXE':
            index_names=['Class']+list(df_tmp.columns[0:12])
            #print(df_tmp.columns)
            df_tmp.set_index(index_names, inplace=True)
            prefixes=['FDR','IncLevel1','IncLevel2','IncLevelDifference']
            filtered_columns = [col for col in df_tmp.columns if any(col.startswith(prefix) for prefix in prefixes)]
            df_filtered = df_tmp[filtered_columns]
            DF=df_filtered
        else:
            df_tmp['None']='None'
            df_tmp['None2']='None'
            index_names=['Class']+list(df_tmp.columns[0:10])+['None']+['None2']
            #print(df_tmp.columns)
            df_tmp.set_index(index_names, inplace=True)
            prefixes=['FDR','IncLevel1','IncLevel2','IncLevelDifference']
            filtered_columns = [col for col in df_tmp.columns if any(col.startswith(prefix) for prefix in prefixes)]
            df_filtered = df_tmp[filtered_columns]
            DF=pd.concat([DF, df_filtered])
        
    # Assuming 'df' is your DataFrame
    DF.index = DF.index.set_names(['Class', 'GeneID', 'geneSymbol', 'chr', 'strand', 'coho1','coho2', 'coho3', 'coho4', 'coho5','coho6', 'coho7', 'coho8'])
    DF.to_csv('merged_DASE/DASE_%s.txt' %(comp),sep='\t')
        
```

```python
LIST_EVENTS=['A3SS', 'A5SS','MXE', 'RI', 'SE']
```

```python
!ls
```

```python
for comp in [comp1, comp2,comp3,comp4,comp5,comp6]:
    denovo=[]
    for clas in LIST_EVENTS:
        novel_path = f'rMATS_{comp}/fromGTF.novelSpliceSite.{clas}.txt'
        # Read the CSV into a DataFrame
        Ndf = pd.read_csv(novel_path, sep='\t', index_col='ID')
        print(novel_path)
        # Collapse values of relevant columns into a single string (use specific columns if needed)
        Ndf['collapsed'] = Ndf.astype(str).agg(' '.join, axis=1)
        denovo=denovo + Ndf['collapsed'].to_list()
        # Convert 'coho7' and 'coho8' to integers, handling NaN values
    DF=pd.read_csv('merged_DASE/DASE_%s.txt' %(comp),sep='\t')
    DF['coho7'] = DF['coho7'].fillna(0).astype(int)  # Fill NaN with 0 and convert to int
    DF['coho8'] = DF['coho8'].fillna(0).astype(int)
    coho=list(DF.columns[1:13])
    # Collapse only the specified columns in 'coho' into a single string
    DF['collapsed'] = DF[coho].apply(lambda row: ' '.join(row.values.astype(str)), axis=1)
    DF['collapsed'] = DF['collapsed'].str.replace(r'(\s?0)+$', '', regex=True)
    DF['DE-NOVO']=False
    # Use pd.Series.isin() to check for matching collapsed strings
    matched_indices = DF[DF['collapsed'].isin(denovo)].index
    # Set the 'DE-NOVO' column to True for all matching rows
    DF.loc[matched_indices, 'DE-NOVO'] = True
    DF=DF.drop(columns=['collapsed'])
    DF[['coho7', 'coho8']] = DF[['coho7', 'coho8']].replace(0, np.nan)
    DF.to_csv('merged_DASE/de_novo_annotated_DASE_%s.txt' %(comp),sep='\t', index=False)
```
